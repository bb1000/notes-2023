## Automatic activation of course environment

The shell is a another name for a computer's command-line interface. There are different shell languages and dialects some more similar than others. Windows has the power shell as well as the older DOS command window.
On Linux and MacOS we have bash and zsh respectively - similar, zsh is newest and back compatible with bash.

Whenever we open a terminal window a shell initialization file is executed. If your default shell is zsh (most macs) verified with

    % ps
        PID TTY          TIME CMD
     147844 pts/10   00:00:00 zsh
     148084 pts/10   00:00:00 ps

the initialization file is `.zshrc` in the home folder. If it is bash (most linux)

    $ ps
        PID TTY          TIME CMD
     147786 pts/10   00:00:00 bash
     148592 pts/10   00:00:00 ps

the initialization file is `.bashrc`


Use a text editor and add the line

    $ conta activate bb1000
    
at the end of the file and this environment will be active whenever you open a new terminal window

## Initialize a project folder

We did

    $ mkdir course-bb1000
    $ cd course-bb1000
    $ git init .
    
to initialize a project folder for git.

## Adding a source file to version control

We created a file savings.py in that folder and typed

    $ git add savings.py
    $ git commit -m "start new project"
    # do some changes
    $ git add savings.py
    $ git commit -m "Initial code..."
    
to save two versions of the code
    
    $ git log --oneline
    a21a8e3 (HEAD -> master) Initial code...
    9a8af43 start new project

* Here `master` is a label for a certain commit, referred to as branch or branch name.
* HEAD is the label for the version that is in the work directory
